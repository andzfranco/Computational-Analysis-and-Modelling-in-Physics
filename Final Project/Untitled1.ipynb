```python
import numpy as np
import matplotlib.pyplot as plt
```

**Generating data points using Numpy array**

```python
x = np.linspace(0, 100, 100)
y = np.linspace(0, 100, 100)
```

**Line plot**

```python
plt.plot(x, y)
```

**Scatter plot**

```python
chocolates = [1, 5, 3, 2, 5, 6, 7]
exam = [78, 93, 90, 85, 92, 99, 100]
```

```python
plt.scatter(chocolates, exam)
```

**Bar plot**

```python
student = ['Joy', 'Goliath', 'Ken', 'Mit']
scores = [95, 87, 95, 99]
```

**Plot customization**

```python
x = np.linspace(0, 100, 10)
y = np.linspace(0, 100, 10)
```

```python
plt.plot(x, y, label = 'line', color = 'purple')
plt.scatter(x, y, label = 'scatter', color = 'red')
plt.title('Customize this plot!', fontsize = '18', fontweight = 'bold', color = 'orange')
plt.xlabel('X-axis', fontsize = '14')
plt.ylabel('y-axis', fontsize = '14')
plt.legend()
plt.grid(alpha = 0.5)
plt.xlim(0, 50)
plt.ylim(0, 50)
plt.savefig('filename.jpg')
```
