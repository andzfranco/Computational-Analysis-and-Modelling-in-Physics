# **Mean Normalized Spectrum**
Before plotting the mean spectrum, we first have to preprocess the data by normalizing it. We can use the `normalize()` function from `sklearn.preprocessing`. Since the data involves spectra of galaxies, the measurements are expressed in lightyears and as a result, the total flux varies largely across the dataset.

````python
from sklearn.preprocessing import normalize

#Normalizing the data
spec_recons = data['spec_recons']
spec_recons = normalize(spec_recons)

#Getting the mean of the data centered at 0 and the standard deviation centered at zero
mu = spec_recons.mean(0)
std = spec_recons.std(0)
````

````python
#Plotting the mean spectrum
plt.plot(wavelength, mu, color = 'lightseagreen', label = 'mean')
plt.fill_between(wavelength, mu - std, mu + std, color='#CCCCCC', label = 'mean $\pm$ std')
plt.title('Mean Normalized Spectrum', weight = 'bold')
plt.xlabel('Wavelength (Angstroms)')
plt.ylabel('Normalized Flux')
plt.legend()
````

The plot above shows the mean spectra of the normalized data. In this plot, we can observe that there are large spikes or variation in some of the wavelengths. To further undertsand what this means, we can compare this mean spectra to the emission atomic spectrum (in Angstroms):

# **Principal Component Analysis**

We've learned that we can omit some of our features, and that in all honesty, dealing with 4000 features and 1000 data points in total will be quite taxing. Because of this, we can actually reduce these features through techniques of dimensionality reduction.

 In this module, we use **Principal Component Analysis (PCA)** to reduce the dimensions in the feature space without compromising the relevant information. While PCA reduces the number of features, it does so while retaining the most informative ones or those with high variance.

Since we want to retain those with high variance, it is important to find the directions of maximum variance in the dataset. These directions are called the **Principal Components**. They are the linear combinations that can maximize the variance of the variables.

The first step in PCA is to obtain the Principal Components. We can use the function `PCA` from `sklearn.decomposition`. We set the number of principal components for this method, in this case, we want the first four (4) principal components.

After obtaining the principal components, we then fit our data and transform them using `fit_transform`.

````python
from sklearn.decomposition import PCA
import pandas as pd

pca = PCA(n_components = 4)
X_proj = pca.fit_transform(spec_recons)
````

To delve deeper into what these components mean, we can show the first few rows of what it comprises of:

````python
component_names = [f'PC{i+1}' for i in range(X_proj.shape[1])]
X_proj_df = pd.DataFrame(X_proj, columns = component_names)
X_proj_df.head()
````

This table shows the scores of the four principals components. These scores reveal the importance or contribution of each data point to the principal components. The sign associated with these scores (positive or negative) indicates how the data aligns with the average data along the direction of the principal component.

# **Eigenspectra of the principal components**

We can plot the eigenvectors of this principal components. Through visualizing the eigenvectors, we can see the major patterns or dimensions along which the data vary.

````python
#Plotting the mean spectrum
plt.plot(wavelength, pca.mean_ - 0.3, color='violet')
plt.text(7000, -0.25, "mean", color='violet')

#Plotting the first 4 eigenvectors
eigenvector_num = 4
for i in range(eigenvector_num):
    color = 'C{}'.format(i)
    plt.plot(wavelength, pca.components_[i] + 0.3 * i, color=color)
    plt.text(7000, -0.01 + 0.3 * i, "component %i" % (i + 1), color=color)

plt.title('Mean spectra and eigenvectors', weight = 'bold')
plt.xlabel('Wavelength (Angstroms)')
plt.ylabel('Normalized flux with offset')
````

The eigenvectors of the principal components tell us about the directions of maximum variance in the data. It also tells us the different properties or variance it can capture. For example, we can see a large spike or variance in component 3, which falls on the wavelength where the hydrogen emission line is located. This suggests that component 3 effectively captures the variability associated with hydrogen emissions.

# **Variance**

To quantify these, we can get the explained variance ratio and the cumulative variance.

The explained variance ratio quantifies how much of the total variance in a dataset is accounted for by each principal component, while the cumulative variance measures how much of the total variability in the dataset is captured with n components.

````python
#We get the number of components, which is 4
n = pca.n_components_
x = np.arange(1, n + 1)

#Getting the explained variance ratio of the n components
evr = pca.explained_variance_ratio_

fig, axs = plt.subplots(1, 2, figsize=(8, 4))

#Plotting the explained variance ratio
axs[0].bar(x, evr, color='lightseagreen')
axs[0].set_xlabel('Component')
axs[0].set_ylabel('Variance ratio')
axs[0].set_title('Explained Variance Ratio', weight = 'bold')
axs[0].set_ylim(0.0, 1.0)
axs[0].set_xticks(range(0, 5, 1))
axs[0].bar_label(axs[0].containers[0])

#Getting the cumulative sum of the explained variance ratio
cv = np.cumsum(evr)

#Plotting the cumulative variance
axs[1].plot(np.r_[0, x], np.r_[0, cv], "o-", color = 'lightseagreen')
axs[1].set_xlabel('Component')
axs[1].set_ylabel('Variance')
axs[1].set_title('Cumulative Variance', weight = 'bold')
axs[1].set_ylim(0.0, 1.0)
axs[1].set_xticks(range(0, 5, 1))

for i, v in enumerate(cv):
    axs[1].text(grid[i], v + 0.03, f"{v:.2f}", ha="center")

plt.tight_layout()
plt.show()
````
