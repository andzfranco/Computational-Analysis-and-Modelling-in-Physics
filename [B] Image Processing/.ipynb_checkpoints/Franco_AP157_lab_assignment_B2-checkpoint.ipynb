For each aperture type, perform the following steps:

  - Generate a 2D array representing a rectangle/circle with the specified size within a square grid of size 128 pixels (representing an optical system, for example).
  
- **Step 2: Compute Fourier Transform**
  - Use the Fast Fourier Transform (FFT) algorithm to compute the Fourier transform of the input aperture.

- **Step 3: Shift the FFT**
  - Shift the zero frequency component of the Fourier transform to the center.

- **Step 4: Visualization (Plotting)**
  - Plot the original aperture.
  - Plot the magnitude of the shifted Fourier transform.

This function transforms an image using FFT for a 2D object and then shifts it to the zero frequency component.

```python
def fft_shift(image):
  fft_shift = fftshift(fft2(image))
  return fft_shift
```

# **Circular apertures**

We define a function that creates circular aperture given the radius and resolution.

```python
def circle_aperture(rad, N):
  xx = np.linspace(-1, 1, N)
  X, Y = np.meshgrid(xx, xx)
  R = np.sqrt(X**2 + Y**2)
  A = np.zeros(np.shape(R))
  A[np.where(R < rad)] = 1.0
  return A
```

```python
#Plotting
cmap = 'magma'
fig, axs = plt.subplots(3, 3, figsize = (8, 8))

#Given radius of the circle apertures
radii = [5, 10, 20]

for i, rad in enumerate(radii):
  N = 128
  circle = circle_aperture(rad/N, N)
  circle_fft = fft_shift(circle)
  circle_logfft = np.log(circle_fft)

  #Plotting
  axs[i, 0].imshow(circle, cmap = cmap)
  axs[i, 0].set_title(f'Radius: {rad} pixels')
  axs[i, 0].axis('off')

  axs[i, 1].imshow(np.abs(circle_fft), cmap = cmap)
  axs[i, 1].set_title(f'Shifted FFT')
  axs[i, 1].axis('off')

  axs[i, 2].imshow(np.abs(circle_logfft), cmap = cmap)
  axs[i, 2].set_title(f'FFT log scale')
  axs[i, 2].axis('off')

plt.tight_layout()
plt.show()
```

As seen from the images above, the shifted FFTs are similar to that of a Bessel function. A larger radii circle results in smaller radius in the shifted FFTm and a smaller radii circle results in a larger shifted FFT. This is correlated to the concept of anamorphism.

Anamorphic means that a signal of N dimension transforms to a 1/N dimension. This means that if a signal appears wide on an axis, it will appear narrow in the spatial frequency axis.

We can relate these circular apertures to that of a camera in a way that when we apply FFT to the camera, we get the point spread function (PSF). The PSF is a measure of the quality of an image, specifically, it describes what a single point in the object looks like in the image.

Source: [Point Spread Function](https://svi.nl/Point-Spread-Function-(PSF#:~:text=In%20fluorescence%20microscopy%2C%20the%20acquired,looks%20like%20in%20the%20image.)

Moreover, the log-scale shifted FFT shows the difference between the values as a significant change in color would be in orders of 10.

# **Rectangular apertures**

We define a function that creates a rectangular aperture given the width, height, and the resolution.

```python
def rect_aperture(width, height, N):
  xx = np.linspace(-1, 1, N)
  X, Y = np.meshgrid(xx, xx)
  A = np.zeros((N,N))
  A[np.where((np.abs(X) <= width/2) & (np.abs(Y) <= height/2))] = 1.0
  return A
```

```python
#Plotting
fig, axs = plt.subplots(2, 3, figsize = (8, 6))
cmap = 'magma'

#Given width and height of rectangular apertures
width_height = [(50, 20), (20, 50)]

for i, (width, height) in enumerate(width_height):
  N = 128
  rectangle = rect_aperture(width/N, height/N, N) #Making a rectangular aperture
  rect_fft = fft_shift(rectangle)
  rect_logfft = np.log(rect_fft)

  #Plotting
  axs[i, 0].imshow(rectangle, cmap = cmap)
  axs[i, 0].set_title(f'{width} x {height} pixels')
  axs[i, 0].axis('off')

  axs[i, 1].imshow(np.abs(rect_fft), cmap = cmap)
  axs[i, 1].set_title(f'Shifted FFT')
  axs[i, 1].axis('off')

  axs[i, 2].imshow(np.abs(rect_logfft), cmap = cmap)
  axs[i, 2].set_title(f'FFT log scale')
  axs[i, 2].axis('off')

plt.tight_layout()
plt.show()
```

For the rectangular apertures, we can observe that their shifted FFTs exhibit anamorphic properties as well. A horizontal rectangle will have strong intensities along the vertical axis, while a vertical rectangle will have strong intensitites along the horizontal axis.

#### PART 2 - Convolution *(20 points)*

Now that we've shown that FT acts as if an imaging system, suppose we have an image and convolve it with a circular aperture with different radii. Assuming that the circular aperture is a camera lens, we want to show how the reconstruction of the original image can be limited by the properties of the camera lens (or any imaging device) such as aperture size.

Take the astronaut image from skimage and convolve it with circular apertures of radii: 8, 32, and 64 pixels.

To do this, you can perform the ff. steps:

- **Compute Fourier Transform of Image:**
  - Use `fft2` to compute the 2D Fourier Transform of the input image.
  
- **Compute Fourier Transform of Aperture:**
  - Use `fftshift` to shift the zero-frequency component to the center.

- **Perform Element-wise Multiplication in Frequency Domain:**
  - Multiply the Fourier Transform of the image and the shifted aperture element-wise in the frequency domain.

- **Compute Inverse Fourier Transform:**
  - Use `ifft2` to compute the 2D Inverse Fourier Transform of the product from the previous step.

- **Visualization (Plotting):**
  - Create a figure with three subplots to visualize the original image, the aperture, and the convolution result.

# **CONVOLUTION**

Convolution is a combination of two signals at different points in time or space. It is defined by the equation:

$f*g(t) = \int_{-\infty}^{\infty} f(\tau) \cdot g(t - \tau) \, d\tau$

In this part, we convolve an image using different sizes of apertures and examine its effect on a resulting image. We do this by multiplying the FT of the image and the aperture, and then taking the inverse of their product.

```python
from skimage import color, data

#Import astronaut image and convert to grayscale
img = color.rgb2gray(data.astronaut())*256

#Given radii
radii = [8, 32, 64]

fig, axs = plt.subplots(3, 3, figsize = (8, 8))

#For each circular aperture
for i, rad in enumerate(radii):
  N = 512
  circle = circle_aperture(rad/N, N) #create a circular aperture using the function used previously
  img_fft = fft2(img)
  aperture_shift = fftshift(circle)
  product = np.multiply(img_fft, aperture_shift)
  inverse = ifft2(product)

  #Plotting
  axs[i, 0].imshow(img, cmap = cmap)
  axs[i, 0].set_title('Original Image')
  axs[i, 0].axis('off')

  axs[i, 1].imshow(circle, cmap = cmap)
  axs[i, 1].set_title('Aperture')
  axs[i, 1].axis('off')

  axs[i, 2].imshow(np.abs(inverse), cmap = cmap)
  axs[i, 2].set_title('Convolution Result')
  axs[i, 2].axis('off')
```

We can see from the above results that a larger aperture results in a more detailed picture of the astronaut. Meanwhile, a smaller aperture results in a blurred image.

A smaller aperture in the Fourier plane makes it a low-pass filter, meaning that it only allows low frequencies to pass through. In optics, a smaller aperture means that only a few angles of light rays can pass. In the fourier plance, this corresponds to filtering out high frequencies. Since high freqeuncies correspond to sharper details and edges, filtering them out makes the image more blurry and having less details. Thus, a smaller aperture results in a stronger convolution.

# **CORRELATION**

Correlation is an image processing technique to see how two signals are closely related. It is defined by the equation:

$(f \star g)(\tau) = \int_{-\infty}^{\infty} f(t) \cdot g(t + \tau) \, dt$

We can use correlation in applications such as template matching, where we have an test/original image and we want to find a certain pattern made by the target image. This is extremelt useful in object detection, character recognition, and fingerprint matching.

Suppose we want to investigate the brush strokes of a painter. However, the texture of the canvas obscures our view of the brushstrokes. Let’s use FFT to remove the canvas weave.

Perform the ff. steps:

1. Open the image AP157_moduleB_canvas_weave_image.jpg (also in the Datasets folder; the image is from a painting by Dr. Vincent Daria.) and convert into a grayscale image. To enhance the detection of repeating patterns we first subtract the mean grayscale from the image. This will remove the DC bias of intensity images.

2. Take the FT of the mean-subtracted image and take note of the symmetric peaks. These are the sinusoids that make up the canvas weave image, similar to the rotation property of sinusoids we did earlier.

3. Manually create a filter mask in the FT space to remove the canvas weave patterns. This is a matrix which is the same size as the image. The filter is all 1’s but are zero at the locations of the sinusoidal peaks. FFT shift this filter and multiply this to the complex FT of the Red, Green and Blue channels of the original image (not mean subtracted). The filter essentially “erases” the peaks. Take the inverse FT of the filtered FT and overlay the filtered R,G,B images. Did the canvas weave vanish enough that the paint and brushstrokes remain?

4. Invert the filter mask (0's become 1's and vice versa) and take the inverse Fourier transform. Observe the generated modulus image. Is it close to the appearance of the canvas weave?

# **Canvas Weave Modeling and Removal**

In this section, we try to remove the canvas the obscures the view in the painting by Dr. Varia. We can do this by first converting the image to grayscale and subtracting the mean grayscale to remove DC bias. We then create a mask that locates the sinusoidal peaks in the image, and set them to a value of 0. This peaks represent the weave pattern that we see in the image. Since we want to clean the image, we want to remove these patterns/peaks, thus why we set it to 0.

 A circular mask was then made and put in the center to ignore the center peak, as this contains the information of the image

Locating the peaks was done through thresholding, depending on the percentile of the brightness of each pixel. If it is above the 99.25 percentile, its value is zero. On the other hand, its value is one. This way, we don't need to manually create shapes and plot them in the peaks.

```python
#Importing the image
canvas_img = cv2.imread(file_path + 'AP157_moduleB2_canvas_weave_image.jpg')
canvas_img = cv2.cvtColor(canvas_img, cv2.COLOR_BGR2RGB)
canvas_gray_img = cv2.cvtColor(canvas_img, cv2.COLOR_RGB2GRAY)

#Subtracting the mean grayscale from the image
canvas_mean_subtracted = canvas_gray_img - np.mean(canvas_gray_img)

#Taking the fourier transform of the mean-subtracted image
canvas_fft = np.abs(np.log(fftshift(fft2(canvas_mean_subtracted))))

#Normalizing
canvas_fft = canvas_fft.astype(float)/np.max(canvas_fft)

#Creating a threshold to mask out the sinusoidal peaks
p1, p99 = np.percentile(canvas_fft, (0, 99.25))
filter = np.where(canvas_fft > p99, 0, canvas_fft) #If the pixel has intensity greater than the 99th percentile, its value is 0

#Creating a circle mask for the center of the image
height, width = 100, 75
x, y = canvas_fft.shape

#Setting the mask value to 1
filter[int((x - width)/2):int((x + width)/2), int((y - height)/2):int((y + height)/2)] = 1

#Shifting the filter
filter_shift = fftshift(np.abs(filter))

#Multiplying thr shifted filter to the Fourier transform of the RGB images
Rd, Gn, Bl = fft2(canvas_img[:, :, 0])*filter_shift, fft2(canvas_img[:, :, 1])*filter_shift, fft2(canvas_img[:, :, 2])*filter_shift

#Getting the inverse of the filtered Fourier transform
invRd, invGn, invBl = np.abs(ifft2(Rd)), np.abs(ifft2(Gn)), np.abs(ifft2(Bl))

#Normalizing the RGB images
norm_inv_R = (invRd - np.min(invRd))/(np.max(invRd) - np.min(invRd))
norm_inv_G = (invGn - np.min(invGn))/(np.max(invGn) - np.min(invGn))
norm_inv_B = (invBl - np.min(invBl))/(np.max(invBl) - np.min(invBl))

masked = np.zeros(canvas_img.shape)

#Overlaying the filtered RGB images
masked[:, :, 0] = norm_inv_R
masked[:, :, 1] = norm_inv_G
masked[:, :, 2] = norm_inv_B

#Plotting
fig, axs = plt.subplots(1, 3, figsize = (16, 10))
cmap = 'gray'
axs[0].imshow(canvas_img, cmap=cmap)
axs[0].set_title('Original Image')
axs[1].imshow(filter, cmap=cmap)
axs[1].set_title('Filter')
axs[2].imshow(masked, cmap=cmap)
axs[2].set_title('Masked Image')
```

We can see from the results above that by applying the mask, the canvas pattern from the image was removed. We can confirm this since we can see how the brush strokes still remained and are more visible.

Now, to further validate and explore, what if we invert the mask. We can see the results below.

```python
#Inverting the filter mask
inv_filter = 1 - np.array(filter)

filter_shift = fftshift(inv_filter)

# Multiplying the shifted filter to the Fourier transform of the RGB images
Rd, Gn, Bl = fft2(canvas_img[:, :, 0]) * filter_shift, fft2(canvas_img[:, :, 1]) * filter_shift, fft2(canvas_img[:, :, 2]) * filter_shift

# Getting the inverse of the filtered Fourier transform
invRd, invGn, invBl = np.abs(ifft2(Rd)), np.abs(ifft2(Gn)), np.abs(ifft2(Bl))

# Normalizing the RGB images
norm_inv_R = (invRd - np.min(invRd)) / (np.max(invRd) - np.min(invRd))
norm_inv_G = (invGn - np.min(invGn)) / (np.max(invGn) - np.min(invGn))
norm_inv_B = (invBl - np.min(invBl)) / (np.max(invBl) - np.min(invBl))

# Overlaying the filtered RGB images
masked = np.zeros(canvas_img.shape)
masked[:, :, 0] = norm_inv_R
masked[:, :, 1] = norm_inv_G
masked[:, :, 2] = norm_inv_B

#Plotting
fig, axs = plt.subplots(1, 3, figsize = (16, 10))
cmap = 'gray'
axs[0].imshow(canvas_img, cmap=cmap)
axs[0].set_title('Original Image')
axs[1].imshow(inv_filter, cmap=cmap)
axs[1].set_title('Inverted filter')
axs[2].imshow(masked, cmap=cmap)
axs[2].set_title('Masked Image')
```

As seen from above, we can see a ragged surface, which resemble the patterns of these canvas weave. Thus, when we use the invert of the filter, we see that the weave patterns are more emphasized and recognizable.

# **Edge Detection**

We have also mentioned that high frequencies can usually represent edges. Thus, we can actually emphasie edges in an image by using a high-pass filter. In this section, we use a high-pass filter on an image to detect its edges. I have here an image that I took at an art museum!




```python
#Importing the image
art = cv2.imread(file_path + 'art.JPG')
art = cv2.cvtColor(art, cv2.COLOR_BGR2RGB)
art_gray = cv2.cvtColor(art, cv2.COLOR_RGB2GRAY)

#Computing the Fourier transform
art_fft = np.abs(np.log(fftshift(fft2(art_gray))))

#Creating a circular mask, high-pass filter
x, y = art_gray.shape
width, height = 50, 50

#Create the filter with all ones
filter = np.ones((x, y), dtype=np.uint8)

#Set the high-pass region/center circle to zero
filter[int((x - width) / 2):int((x + width) / 2), int((y - height) / 2):int((y + height) / 2)] = 0

#Applying the high-pass filter
art_filtered = ifftshift(fftshift(fft2(art_gray)) * filter)
art_filtered = np.abs(ifft2(art_filtered))

#Plotting
fig, axs = plt.subplots(1, 4, figsize=(16, 10))
axs[0].imshow(art_gray, cmap='gray')
axs[0].set_title('Original')
axs[0].axis('off')
axs[1].imshow(art_fft, cmap='gray')
axs[1].set_title('Fourier Transform')
axs[1].axis('off')
axs[2].imshow(filter, cmap='gray')
axs[2].set_title('High-pass filter')
axs[2].axis('off')
axs[3].imshow(art_filtered, cmap='gray')
axs[3].set_title('Edge detected using HPF')
axs[3].axis('off')
plt.show()
```

From the results, we can see that the high-pass filter is just like a little dot. This is because we set the low frequencies to 0, thus making only the center circle with high frequency to remain. Moreover, the final image shows us outlines of the art, in particular, the shapes outline, the eyes, and the black dots at the background.
